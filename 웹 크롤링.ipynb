```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
driver=webdriver.Chrome(ChromeDriverManager().install())
```

```python
url="https://www.melon.com/chart/index.htm"
```

```python
driver.get(url)
```

```python
from bs4 import BeautifulSoup
```

```python
html=driver.page_source
soup=BeautifulSoup(html,'html.parser')
```

```python
soup
```

```python
songs=soup.select("table > tbody >tr")
```

```python
songs
```

```python
for song in songs:
    b=song.select("span > a")
    title=b[0].string
    singer=b[1].string
    print(title,singer)
    
```

```python
song_data=[]
for song in songs:
    title=song.select_one("div.rank01").text
    singer=song.select_one("div.rank02 > a").string
    song_data.append((title,singer))
print(song_data)
```

```python
import pandas as pd
```

```python
pd.DataFrame(song_data)
melon_df=pd.DataFrame(song_data, columns=['제목','가수']) 
```

```python
melon_df
```

```python
url="https://www.genie.co.kr/chart/top200"
```

```python
driver.get(url)
```

```python
html=driver.page_source
soup=BeautifulSoup(html,'html.parser')
```

```python
soup
```

```python
songs=soup.select("table > tbody> tr")
```

```python
song_data=[]
rank=0
for song in songs:
    rank+=1
    b=song.select("td.info > a")
    title=b[0].string.strip()
    singer=b[1].string.strip()
    song_data.append([rank,title,singer])
print(song_data)
```

```python
data=pd.DataFrame(song_data, columns=['순위','제목','가수'])
```

```python
song_data=[]
rank=1
for song in songs:
    title=song.select_one("a.title").string.strip()
    singer=song.select_one("a.artist").string.strip()
    song_data.append(['Ginie',rank,title,singer])
    rank+=1
```

```python
genie_df=pd.DataFrame(song_data, columns=['서비스','순위','타이틀','가수'])
genie_df
genie_df.to_excel("genie.xlsx",index=False)
```
